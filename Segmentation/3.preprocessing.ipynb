```python
import os
from python_speech_features import mfcc
import scipy.io.wavfile as wav
import pandas as pd
import pickle
import math
```

```python
max_frames=0
speakers = set()
```

```python
def save_file(files_wav,files_text,filename,max_frames,speakers):
    """Extracting MFCC Features and ignoring files with audio greater than 5 sec"""
    dataset = []
    for i in range(len(files_wav)):  ###############
        flag = False
        speaker_id = int(files_wav[i].split("/")[-2][1:])
        reader = open(files_text[i],"r")
        text = reader.readline()
        (rate,sig) = wav.read(files_wav[i])
        mfcc_feat = mfcc(sig,rate, numcep = 13, nfilt =26 ,nfft=1200) #########
        if(len(mfcc_feat)<=500):
            flag = True
            speakers.add(speaker_id)
            max_frames = max(max_frames, len(mfcc_feat))
            dataset.append([speaker_id,text,mfcc_feat,len(mfcc_feat),files_wav[i]]) #storing data as table of [text , mfcc_features] pairs
        reader.close()
        print i, flag
    with open("../dataset/"+ filename,"w") as f: ########
        pickle.dump(dataset,f)
        print "saved"
    print dataset[0]
    return [max_frames,speakers]    
```

```python
"""getting directories for wav and corresponding text file"""
os.chdir("/mnt/Ext1/TTS/Segmentation/VCTK-Corpus") #path for the dataset goes here
wav_dir = (os.getcwd()+'/wav48')
text_dir = (os.getcwd()+'/txt')
```

```python
"""getting all the directories containing wav files"""
directories = []
for dirpaths,dirnames,filenames in os.walk(wav_dir):
    directories.append(dirpaths)
directories = directories[1:] #ignoring current directory
```

```python
"""creating list of paths for all wav files and corresponding path for text file in the dataset"""
files_wav = []
files_text = []
for i in directories:
    for dirpaths,dirnames,filenames in os.walk(i):
        for j in filenames:
            files_wav.append(i+'/'+j.split(".")[0]+".wav")
            #print (i+'/'+j.split(".")[0]+".wav")
            files_text.append(i.replace("wav48","txt")+'/'+j.split(".")[0]+".txt")
            #print (i.replace("wav48","txt")+'/'+j.split(".")[0]+".txt")
```

```python
"""Pickling MFCC Features for audio files
    Change batch size for increasing number of files for effective RAM usage"""
batch_size = 1600000  ##infinity
files_selected = files_wav[:]
count = int(math.ceil(float(len(files_selected))/float(batch_size))) ###########
for i in range(count):
    start = i*batch_size
    end = min((i+1)*batch_size,len(files_selected))
    max_frames, speakers = save_file(files_wav[start:end],files_text[start:end],str(i+1)+".pkl",max_frames,speakers)
```

```python
"""Storing a list of Speakers along with the maximum number of frames and the number of files for use in Segmentation Model"""
speakers = list(speakers)

#save dataset to file
with open("../dataset/max_frames_speakers_count.pkl","w") as f: ########
    pickle.dump([max_frames,speakers,count],f)
    print "saved"

print "max frames: "+ str(max_frames)
print speakers
print count
```
